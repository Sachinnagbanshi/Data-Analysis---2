```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
holiday_df=pd.read_csv('US_Holiday_Dates_(2004-2021).csv')
holiday_df.sample(10)
```

```python
ecom_df=pd.read_csv('Ecommerce_Data.csv')
ecom_df.sample(5)
```

```python
holiday_df.info()
```

```python
#converting the date column from string to date-time format

holiday_df.Date=pd.to_datetime(holiday_df.Date)
ecom_df.Date=pd.to_datetime(ecom_df.Date)
```

```python
holiday_df.info()
```

```python
ecom_df.info()
```

```python
#joining both the tables in terms of the date

df=ecom_df.join(holiday_df,lsuffix='Date',rsuffix='Date')
```

```python
df.isna().mean()*100
```

```python
#renaming column

df.rename({'key_0':'Date'},inplace=True,axis=1)
```

```python
df.columns
```

```python
#droping unnamed column

df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.Holiday.value_counts()
```

```python
#Creating a sales column, (Sales=Qunatity * UnitPrice)

df['Sales']=df['Quantity']*df['UnitPrice']
```

```python
df
```

```python
Total_Sales=df.groupby(['Holiday'])['Sales'].agg(['sum']).reset_index().sort_values('sum',ascending=False)
Total_Sales.rename({'sum':'Total_sale'},inplace=True,axis=1)
plt.figure(figsize=(18,9),dpi=100)
sns.set_theme(context='poster',style='whitegrid',palette='bright')
sns.barplot(Total_Sales.Total_sale,Total_Sales.Holiday)
Total_Sales
```

# So here we can note the Total Sales are comparitively higher during Labor Day Weekend with 8.95 % of the Total sales amount which is 575 

```python
Avg_Sales=df.groupby(['Holiday'])['Sales'].agg(['mean']).reset_index().sort_values('mean',ascending=False)
Avg_Sales.rename({'mean':'Avg_sale'},inplace=True,axis=1)
sns.set_theme(context='poster',style='whitegrid',palette='bright')
fig=plt.figure(figsize=(16,8),dpi=100)
sns.barplot(Avg_Sales.Avg_sale,Avg_Sales.Holiday)
Avg_Sales['perc']=(Avg_Sales['Avg_sale']/Avg_Sales['Avg_sale'].sum())*100
Avg_Sales
```

# Here the above bar chart shows the Average sales during all Holidays, During the Washington's Birthday the Average sales is maximum with 31.59 amount sales per day.

```python
a=df.drop(df.loc[df.Holiday=="Washington's Birthday"].index)
df.Sales.sum()-a.Sales.sum()
```

# As we can notice here the Sales is most positively correlated to Quantity so as we increase the product quantity the sales would eventually increase,Also the sales is most negatively correlated to the Unit price, which means as the Unit price increases the sales eventually decreases.

```python
Avg_Sales=df.groupby(['WeekDay'])['Sales'].agg(['mean']).reset_index().sort_values('mean',ascending=False)
Avg_Sales.rename({'mean':'Avg_sale'},inplace=True,axis=1)
sns.set_theme(context='poster',style='whitegrid',palette='bright')
plt.figure(figsize=(16,8))
plt.pie(Avg_Sales.Avg_sale,labels=Avg_Sales.WeekDay,autopct='%0.2f')
Avg_Sales
```

# Fridays produce maximum no. of sales per unit and Thursdays produce minimum sales.

```python
fig=plt.figure(figsize=(14,7),dpi=100)
Avg_Sales=df.groupby(['Country'])['Sales'].agg(['mean']).reset_index().sort_values('mean',ascending=False)
Avg_Sales.rename({'mean':'Avg_sale'},inplace=True,axis=1)
sns.set_theme(context='notebook',style='whitegrid',palette='bright',font_scale=1)
sns.barplot(Avg_Sales.Avg_sale,Avg_Sales.Country,palette='summer',errwidth=0.1,saturation=1)
Avg_Sales['perc']=(Avg_Sales['Avg_sale']/Avg_Sales['Avg_sale'].sum())*100
Avg_Sales
```

```python
plt.figure(figsize=(12,6))
Avg_Sales=df.groupby(['Hour'])['Sales'].agg(['mean']).reset_index().sort_values('mean',ascending=False)
Avg_Sales.rename({'mean':'Avg_sale'},inplace=True,axis=1)
sns.set_theme(context='notebook',style='whitegrid',palette='bright',font_scale=1)
sns.barplot(Avg_Sales.Hour,Avg_Sales.Avg_sale)
Avg_Sales['perc']=(Avg_Sales['Avg_sale']/Avg_Sales['Avg_sale'].sum())*100
Avg_Sales
```
